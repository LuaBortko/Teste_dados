## Tratando os dados 
### Nulos
#### Substituir os dados faltantes com a média ou mediana

```python
df_salarios = pd.DataFrame({
    "nome": ["Ana","Bruno","Carlos","Danielle","Val"],
    "salario":[4000,np.nan,5000,np.nan,100000]
})
#Criando um dataFrame para testes
df_salarios["salario_media"] = df_salarios["salario"].fillna(df_salarios["salario"].mean().round(2))
#fillna substitui os valores nulos com o que ta no parenteses / round arredonda
df_salarios["salario_mediana"] = df_salarios["salario"].fillna(df_salarios["salario"].median())
#Substituindo pela mediana (as vezes melhor quando temos valores muito discrepantes nas pontas)
df_salarios
```

#### Substituir com valores anteriores

```python
df_temperatura = pd.DataFrame({
    "dia": ["segunda","terça","quarta","quinta","sexta"],
    "temperatura": [30,np.nan,np.nan,28,27]
})
df_temperatura["preechido_ffill"] = df_temperatura["temperatura"].ffill()
#Substitui os nulos com o valor anterior
df_temperatura["preechido_bfill"] = df_temperatura["temperatura"].bfill()
#Substitui os nulos com o valor depois
df_temperatura
```

#### Substituir por não informado

```python
df_cidades = pd.DataFrame({
    "nome": ["Ana","Bruno","Carlos","Danielle","Val"],
    "cidade": ["São Paulo", np.nan, "Curitiba",np.nan,"Belem"]
})
df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("não informado")
#fillna substitui os valores nulos com o que ta no parenteses
df_cidades
```
