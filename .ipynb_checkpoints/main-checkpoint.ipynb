```python
df.columns #lista com o nome das colunas
```

```python
# Dicionário de renomeação
novos_nomes = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}
# Aplicando renomeação
df.rename(columns=novos_nomes, inplace=True)
df.columns
```

```python
df["senioridade"].value_counts()
#ver a frequencia dos valores de uma coluna expecifica
```

```python
df["contrato"].value_counts()
```

```python
#substituindo as siglas
senioridade = {
    'SE': 'senior',
    'MI': 'pleno',
    'EN': 'junior',
    'EX': 'executivo'
}
contrato = {
    'FT': 'integral',
    'PT': 'parcial',
    'CT': 'contrato',
    'FL': 'freelancer'
}
tamanho_empresa = {
    'L': 'grande',
    'S': 'pequena',
    'M':	'media'
}
mapa_trabalho = {
    0: 'presencial',
    100: 'remoto',
    50: 'hibrido'
}
df['senioridade'] = df['senioridade'].replace(senioridade)
df['contrato'] = df['contrato'].replace(contrato)
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
df['remoto'] = df['remoto'].replace(mapa_trabalho)
df['senioridade'].value_counts()
```

```python
df.head(20)
```

```python
df.describe(include="object")
#Resumo da tabela, mas incluindo os objetos
#top é a informação mais frequente -> moda,  e freq o valor de vezes que aparece a informação mais frequente 
```
